```python
import requests
from bs4 import BeautifulSoup
```

```python
## 주소만들기
url = 'https://book.naver.com/search/search.nhn'
params = {'sm':'sta_hty.book','sug':'','where':'nexearch','query':'bigdata'}
```

```python
## get요청
response = requests.get(url,params=params)
# status_code = response.status_code
# status_code
text = response.text
print(text)
```

```python
## str => bs객체
soup= BeautifulSoup(text)
```

```python
## 책 전체 정보 가져오기 => 크롬 개발자 도구
book_all = soup.find(id='searchBiblioList')
#book_all = soup.select('#searchBiblioList')
book_all
```

```python
book_all_li_all=book_all.select('li')   ## 책전체
book_one_li_one = book_all.select_one('li')   ##책 한개
```

```python
## 책 bid 가져오기

bid_list=[]
book_all_li_one.a['href'].split('=')[1]

for book_all_li_one in book_all_li_all:
    b_id = book_all_li_one.a['href'].split('=')[1]
    bid_list.append(b_id)
bid_list
```

```python
## 책 제목 가져오기

title_list = [book.img['alt'] for book in book_all_li_all]
title_list
```

```python
## 저자, 출판사, 출판일, 가격(정가-할인가) 가져오기
author_list = []
cps_list = []
date_list = []
price1 = []
price2=[]
```

```python
### 강사님 풀이

book_text = book_all_li_one.select_one('dd.txt_block').text
book_text = book_text.replace('\n','').replace('\xa0','').replace('\t','').replace('\r','')
book_text_list = book_text.split('|')
book_author = book_text_list[0]
book_publish = book_text_list[1]
book_pubdate = book_text_list[2]
print(book_author, book_publish, book_pubdate)
```

```python
### 강사님 풀이

auth_li = []
pub_li=[]
pub_date=[]


for book_all_li_one in book_all_li_all:
    book_text = book_all_li_one.select_one('dd.txt_block').text
    book_text = book_text.replace('\n','').replace('\xa0','').replace('\t','').replace('\r','')
    book_text_list = book_text.split('|')
    if len(book_text_list)==4:
        book_author = book_text_list[0] + book_text_list[1]
        book_publish = book_text_list[2]
        book_pubdate = book_text_list[3]
    else:
        book_author = book_text_list[0]
        book_publish = book_text_list[1]
        book_pubdate = book_text_list[2]
    auth_li.append(book_author)
    pub_li.append(book_publish)
    pub_date.append(book_pubdate)
print(auth_li)

```

```python
### 강사님 풀이

book_desc = book_all_li_all[4].select_one('dd.txt_desc')
price_old = book_desc.select_one('strike')
price_new = book_desc.select_one('em.price')
if price_old == None:
    price_old = 0
else:
    price_old = price_old.split('원')[0]
if price_new == None:
    price_new = 0
else:
    price_new = price_old.split('원')[0]
    print(price_old, price_new)
# price_new = book_desc.select_one('em.price').text
# price_new = price_new.split('원')[0]
# print(price_old, price_new)
```

```python
### 강사님 풀이

prices_list = []

for book_all_li_one in book_all_li_all:
    book_desc = book_all_li_one.select_one('dd.txt_desc')
    price_old = book_desc.select_one('strike')
    price_new = book_desc.select_one('em.price')
    if not price_old:
        price_old = 0
    else:
        price_old = price_old.text.split('원')[0]
    if not price_new:
        price_new = 0
    else:
        price_new = price_new.text.split('원')[0]
    prices_list.append((price_old, price_new))
prices_list
```

```python
### 강사님 풀이

book_info_list = []

for i in range(len(bid_list)):
    book_info_dict = {}
    book_info_dict['bid'] = bid_list[i]
    book_info_dict['title'] = title_list[i]
    book_info_dict['author'] = auth_li[i]
    book_info_dict['publish'] = pub_li[i]
    book_info_dict['pubdate'] = pub_date[i]
    book_info_dict['price_old_new'] = prices_list[i]
    book_info_list.append(book_info_dict)
    


```
