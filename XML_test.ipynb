## 기상청

```python
import urllib.request as req
from bs4 import BeautifulSoup
import os.path
import time
```

```python
url = 'http://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
dt = time.strftime('%Y%m%d',time.localtime(time.time()))
xml_file = f'forecast_{dt}.xml'  ## 파일명
```

```python
## xml 파일로 저장
if not os.path.exists(xml_file):
    req.urlretrieve(url, xml_file)

xml = open(xml_file,'r',encoding='utf-8').read()
soup = BeautifulSoup(xml)
```

```python
locations = soup.select('location')
locations
```

```python
location = soup.select_one('city').text
location
```

```python
#city=[]
info={}
info

for location in locations:
    city=location.select_one('city').text
    weather = location.select_one('wf').text
    if not (weather in info):
        info[weather]=[]
    info[weather].append(city)

info
```

```python
## 서울경기
url='http://www.kweather.co.kr/data/JISU/11B00000.xml'
xml_file = 'kweather_seoul.xml'
```

```python
if not os.path.exists(xml_file):
    req.urlretrieve(url, xml_file)

xml = open(xml_file,'r',encoding='utf-8').read()
soup = BeautifulSoup(xml)
soup
```

```python
nums={}
num=soup.select('jtitle')

for i in num:
    n = i.select_one('jnum').text
    text = i.select_one('jtext').text
    if not (text in nums):
        nums[text] = []
    nums[text].append(n)
```
