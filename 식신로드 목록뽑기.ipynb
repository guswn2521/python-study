```python
import requests
from bs4 import BeautifulSoup
import folium
```

```python
url = 'https://www.wikitree.co.kr/articles/217101'
response = requests.get(url)
soup = BeautifulSoup(response.text)
ps = soup.select('div p')
name= soup.select('div strong')  ##31(index 0~30)
```

```python
## 상호 크롤링

temp_name=[]

for i in range(len(name)):
    if i<=27:
        a = name[i].text.split('회')
        temp_name.append(a[-1])
    elif i==29:
        temp_name.append(name[i].text.split()[-1])

for i in temp_name:
    if len(i)==0:
        temp_name.remove(i)
    if '\xa0' in i:
        temp_name.remove(i)
temp_name=[name.replace(' ','') for name in temp_name if len(name) != 0]
temp_name
```

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.wikitree.co.kr/articles/217101'
response = requests.get(url)
resp = response.text.replace('&nbsp;','')
soup = BeautifulSoup(resp)
soup
```

```python
raw_data = soup.select_one('div#wikicon')
raw_data
```

```python
title_one = raw_data.select_one('div strong')
title_one
```

```python
title_all = raw_data.select('div strong')
first_title_list = []
for title in title_all:
    title_text = title.text
    if title_text:
        if title_text.strip() == '14.' or title_text.strip() == '20.':
            continue
        else:
            first_title_list.append(title_text)
        
len(first_title_list)
```

```python
second_title_list=[]
for i in first_title_list:
    second_title_list.append(i.split('회 ')[1])
title_list = second_title_list
```

```python
tel_p = re.compile('\d{2,3}-\d{3,4}-\d{4}')
tel_list=[]
tel_all = raw_data.select('p')
for item in tel_all:
    item = str(item)
    info = tel_p.findall(item)
    if info:
        tel_list.append(info)
len(tel_list)
```

```python
## 전화번호 찾기 정규표현식
tel_p = re.compile('\d{2,3}-\d{3,4}-\d{4}')
tel_data=raw_data.find_all(string = tel_p)
tel_list=[]
for tel_one in tel_data:
    tel_list.append(tel_one.strip())
print(len(tel_list))
```

```python
## 주소찾기
address_p = re.compile('^.*구\s.*동')
address_list=raw_data.find_all(string = address_p)
print(address_list)
```

```python
restaurant_list = []
for i in range(len(title_list)):
    info = (title_list[i], tel_list[i], address_list[i])
    restaurant_list.append(info)
```
