```python
## url 만들기

url = 'https://book.naver.com/search/search.nhn'

#파라메터
values={'sm':'sta_hty.book', 'sug':'','where':'nexearch','query':'파이썬'}
#딕셔너리->쿼리스트링
params=urllib.parse.urlencode(values)

url = url+'?'+params
print(url)
```

```python
## 웹에서 데이터 읽어오기

response = urllib.request.urlopen(url)  ## http요청하고 결과를 받아옴
# status_code = response.getcode()
# print(status_code)
data = response.read()
text = data.decode('utf-8')
print(text)
```

### 2. header 값 추가하기

```python
## PC

url = 'http://www.naver.com'
header_info = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36'}
request = urllib.request.Request(url, headers=header_info)

resp = urllib.request.urlopen(request)
print(resp.geturl())
print(resp.info())
```

```python
## 모바일
url = 'http://www.naver.com'
header_info = {'user-agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Mobile Safari/537.36'}
request = urllib.request.Request(url, headers=header_info)

resp = urllib.request.urlopen(request)
print(resp.geturl())
print(resp.info())
```

### 3. requests 모듈
- .get() : 요청 , headers   => get요청 : params, post요청 : data
- .status_code
- .text : 응답내용 읽음

```python
import requests
url = 'https://book.naver.com/search/search.nhn'
#파라메터
params={'sm':'sta_hty.book', 'sug':'','where':'nexearch','query':'파이썬'}
```

```python
# Get 요청 : params 사용

resp = requests.get(url, params = params)            ## params에 데이터 넘김 (url을 만듬)
status_code = resp.status_code
data = resp.text                 ## type() : str
print(status_code)
print(data)  
```

```python
# Post 요청 : data사용

resp = requests.get(url, data = params)
status_code = resp.status_code
print(status_code)
data = resp.text
print(data)
```

```python
# from bs4 import BeautifulSoup
# soup = BeautifulSoup(data, 'html.parser')  ## data에 든 자료를 str -> bs4.BeautifulSoup 객체
# print(soup.find('h3'))
## post방식으로 데이터 검색 안된다는데 뭔소린지모르겠음
```

```python
#header 추가 : headers 사용

#모바일 헤더
header_info = {'user-agent':'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Mobile Safari/537.36'}
#get(url, GET, HEADER)
resp = requests.get(url, headers = header_info, params = params)
#resp = requests.get(url,headers='',params = params)

status_code = resp.status_code
if status_code==200:
    data = resp.text
    resp_url = resp.url
    print(resp_url)
    print(data)

```
