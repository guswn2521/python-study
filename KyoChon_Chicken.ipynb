```python
def get_requests(i,j):
    url = 'http://www.kyochon.com/shop/domestic.asp'
    params = {'sido1':i,'sido2':j,'txtsearch':''}
    response = requests.get(url,params=params)
    #global cnt
    if response.status_code==200:
        return response.text
    else:
        print(response.status_code)
```

```python
addr=[];jijum=[];x=[];y=[];tel=[]
def append_list(i,j):
    response = get_requests(i,j)
    soup = BeautifulSoup(response)
    shops = soup.select_one('ul.list').select('li')
    tels = soup.select('span.store_item em')
    for shop in shops:
        info = shop.select_one('a')['href'].split("'")
        addr.append(info[1])
        jijum.append(info[3])
        x.append(info[-4])
        y.append(info[-2])
    for item in tels:
        a = item.text.split('\r\n\t\t\t\t\t\t\t\t\t\t')
        tel_one = a[-1].strip('\r\n\t\t\t\t\t\t\t\t\t')
        tel.append(tel_one)
```

```python

sido = [i for i in range(1,18)]
sido2 = [j for j in range(1,26)]

for i in sido:
    for j in sido2:
        append_list(i,j)
```
