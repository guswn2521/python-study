```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
params = {'sel':'pnt','date':'20200603'}
```

```python
response = requests.get(url,params=params)
data = response.text
print(data)
```

```python
soup = BeautifulSoup(data)
```

```python
movies = soup.tbody.select('tr')
movie = movies[1]
movies
```

```python
movie.a['title']
```

```python
movie_titles = []
for movie in movies:
    if movie.a == None:
        pass
    else:
        movie_titles.append(movie.a['title'])
print(movie_titles)
```

```python
points = []
#print(movie.select_one('td.point').text)
for movie in movies:
    a = movie.select_one('td.point')
    if a ==None:
        pass
    else:
        a = a.text
        points.append(a)
print(points)
```

```python
movie_ranking=[]
for i in range(len(movie_titles)):
    dic={}
    dic[movie_titles[i]]= points[i]
    movie_ranking.append(dic)
print(movie_ranking)
    
    
```

```python
import requests
from bs4 import BeautifulSoup
import datetime
```

```python
def get_naver_movie(sel = 'cur', date = '20200604'):
    url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
    params = {'sel':sel,'date':date}
    response = requests.get(url, params = params)
    status_code = response.status_code
    if status_code ==200:
        text=response.text
        soup = BeautifulSoup(text)
    else:
        soup = 'error'
    return soup
```

```python
def get_date(day):
    now = datetime.datetime.now()
#     tomorrow = now+datetime.timedelta(days=1)
#     yesterday = now+datetime.timedelta(days=-1)
    ## date object => str
    #yesterday.strftime('%Y%m%d')
    ret_date = now +datetime.timedelta(days=day)
    return ret_date.strftime('%Y%m%d')
```

```python
p_date = get_date(-10)
p_date
```

```python
soup = get_naver_movie()
soup
```

```python
movie_table = soup.select_one('table.list_ranking')
movie_tr_all = movie_table.select('tr')
movie_tr_all[2]
```

```python
movie_tr_one = movie_tr_all[2]
movie_title = movie_tr_one.select_one('a').text
movie_point = movie_tr_one.select_one('td.point').text
print(movie_title, movie_point)
```

```python
movie_title_point_list=[]
for movie_tr_one in movie_tr_all:
    movie_title = movie_tr_one.select_one('a')
    movie_point = movie_tr_one.select_one('td.point')
    if movie_title:
        title = movie_title.text
        point = movie_point.text
    else:
        continue
    movie_title_point_list.append((title, point))
movie_title_point_list
```
